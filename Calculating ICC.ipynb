```python
columnlist = pd.Index.tolist(features.columns)
print(columnlist)
toremove = ["Unnamed: 0", "ROI",'diagnostics_Versions_PyRadiomics', 'diagnostics_Versions_Numpy', 'diagnostics_Versions_SimpleITK', 'diagnostics_Versions_PyWavelet', 'diagnostics_Versions_Python', 'diagnostics_Configuration_Settings', 'diagnostics_Configuration_EnabledImageTypes', 'diagnostics_Image-original_Hash', 'diagnostics_Image-original_Dimensionality', 'diagnostics_Image-original_Spacing', 'diagnostics_Image-original_Size', 'diagnostics_Image-original_Mean', 'diagnostics_Image-original_Minimum', 'diagnostics_Image-original_Maximum', 'diagnostics_Mask-original_Hash', 'diagnostics_Mask-original_Spacing', 'diagnostics_Mask-original_Size', 'diagnostics_Mask-original_BoundingBox', 'diagnostics_Mask-original_VoxelNum', 'diagnostics_Mask-original_VolumeNum', 'diagnostics_Mask-original_CenterOfMassIndex', 'diagnostics_Mask-original_CenterOfMass', 'Patient']
new_list = []
featurelist = []
for feature in columnlist:
    if feature not in toremove:
        featurelist.append(feature)

print(featurelist)
print(type(featurelist))
```

```python
for targetfeature in featurelist:
    lesion = features["ROI"]
    lesion = pd.Series.to_frame(lesion)
    feature = features[targetfeature]
    lesion = lesion.join(feature)
    lesion[''] = targetfeature
    lesion.rename(mapper=None, index=None, columns = {'':'feature'}, inplace = True)
    lesion.rename(mapper=None, index=None, columns = {targetfeature:'value'}, inplace = True)
    target_dataframe = r"E:\Internship Mass storage\First 3 datasets\VALDO dataset\Dataframes for ICC/" + targetfeature + ".xlsx"
    lesion.to_excel(target_dataframe, sheet_name="Sheet1")
```

```python
for targetfeature in featurelist:
    featuredataframe = r"E:\Internship Mass storage\First 3 datasets\VALDO dataset\Dataframes for ICC/" + targetfeature + ".xlsx"
    iccdataframe = pd.read_excel(featuredataframe, sheet_name="Sheet1")
    icc = pg.intraclass_corr(data=iccdataframe, targets='', raters='ROI', ratings=targetfeature)
```
